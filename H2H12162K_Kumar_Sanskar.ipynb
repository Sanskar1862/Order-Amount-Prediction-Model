### 1.Check for the outliers in the “amount_in_usd” column and replace the outliers with appropriate values, discussed in the sessions.

```python
from sklearn.feature_selection import SelectKBest
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
```

```python
data = pd.read_csv("Final1.csv")
```

```python
median = data["amount_in_usd"].median()
print(median)
```

```python
Q1 = data["amount_in_usd"].quantile(0.25)
Q3 = data["amount_in_usd"].quantile(0.75)
iqr = Q3 - Q1
print(iqr)
```

```python
threshold = 1.5 * iqr
```

```python
outliers = (data["amount_in_usd"] < Q1 - threshold) | (data["amount_in_usd"] > Q3 + threshold)
print(len(outliers))
```

```python
lower = Q1 - threshold
upper = Q3 + threshold
```

```python
data["amount_in_usd"] = data["amount_in_usd"].mask(data["amount_in_usd"]>upper , median)
data["amount_in_usd"] = data["amount_in_usd"].mask(data["amount_in_usd"]<lower , median)
```

# Milestone 4 ML Models and Evaluations

### 1. Modify the dataset to pass into any type of machine learning models.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
%matplotlib inline
```

```python
columns_to_drop = ['ORDER_AMOUNT']
for i in columns_to_drop:
    data = data.drop(i,axis = 1)
```

```python
data.rename(columns = {'amount_in_usd': 'target'}, inplace = True)
```

```python
data.dropna(inplace = True)

X = data.drop(columns = ['target'])
y = data['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 469)
```

### 2.Try different machine learning models like -
    Linear Regression
    Support Vector Machine
    Decision Tree
    Random Forest
    AdaBoost
    Xgboost etc.

```python
from sklearn.linear_model import LinearRegression
# Linear Regression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
# Initialize the models
linear_reg = LinearRegression()
# svm = SVR()
decision_tree = DecisionTreeRegressor()
random_forest = RandomForestRegressor()
adaboost = AdaBoostRegressor()
xgboost = xgb.XGBRegressor()

# Fit the models
linear_reg.fit(X_train, y_train)
# svm.fit(X_train, y_train)
decision_tree.fit(X_train, y_train)
random_forest.fit(X_train, y_train)
adaboost.fit(X_train, y_train)
xgboost.fit(X_train, y_train)

# Make predictions on the test set
linear_reg_preds = linear_reg.predict(X_test)
# svm_preds = svm.predict(X_test)
decision_tree_preds = decision_tree.predict(X_test)
random_forest_preds = random_forest.predict(X_test)
adaboost_preds = adaboost.predict(X_test)
xgboost_preds = xgboost.predict(X_test)

# Evaluate the models
linear_reg_mse = mean_squared_error(y_test, linear_reg_preds)
# svm_mse = mean_squared_error(y_test, svm_preds)
decision_tree_mse = mean_squared_error(y_test, decision_tree_preds)
random_forest_mse = mean_squared_error(y_test, random_forest_preds)
adaboost_mse = mean_squared_error(y_test, adaboost_preds)
xgboost_mse = mean_squared_error(y_test, xgboost_preds)

linear_reg_rmse = mean_squared_error(y_test, linear_reg_preds, squared=False)
# svm_rmse = mean_squared_error(y_test, svm_preds, squared=False)
decision_tree_rmse = mean_squared_error(y_test, decision_tree_preds, squared=False)
random_forest_rmse = mean_squared_error(y_test, random_forest_preds, squared=False)
adaboost_rmse = mean_squared_error(y_test, adaboost_preds, squared=False)
xgboost_rmse = mean_squared_error(y_test, xgboost_preds, squared=False)

linear_reg_r2 = r2_score(y_test, linear_reg_preds)
# svm_r2 = r2_score(y_test, svm_preds)
decision_tree_r2 = r2_score(y_test, decision_tree_preds)
random_forest_r2 = r2_score(y_test, random_forest_preds)
adaboost_r2 = r2_score(y_test, adaboost_preds)
xgboost_r2 = r2_score(y_test, xgboost_preds)
```

### 3.Perform Regression model evaluations like MSE, RMSE, R-Square etc.

```python
y_pred = regressor.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Linear Regression Model:")
print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mae)
print("R-Squared:", r2)
```

### 4.Compare the accuracies of all the models 

```python
# Compare the performances
model_performances = pd.DataFrame({
    'Model': ['Linear Regression','Decision Tree', 'Random Forest', 'AdaBoost', 'XGBoost'],
    'MSE': [linear_reg_mse,decision_tree_mse, random_forest_mse, adaboost_mse, xgboost_mse],
    'RMSE': [linear_reg_rmse,decision_tree_rmse, random_forest_rmse, adaboost_rmse, xgboost_rmse],
    'R-Squared': [linear_reg_r2,decision_tree_r2, random_forest_r2, adaboost_r2, xgboost_r2]
})

print(model_performances)
```

### 5.Select the best possible model
